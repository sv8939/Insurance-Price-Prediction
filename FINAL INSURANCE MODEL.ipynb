```python
gender = {"male":0,"female":1}
df['sex'] = df['sex'].map(gender)
df
```

```python
df1 = {"yes":1,"no":0}
df['smoker'] = df['smoker'].map(df1)
df
```

```python
unique_values = df['region'].unique()
unique_values
```

```python
df1 = {"southwest":1,"southeast":0,"northwest":2,"northeast":3}
df['region'] = df['region'].map(df1)
df
```

```python
X = df.drop(df.columns[[6]], axis=1)
Y=df['charges']
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
Y_pred = model.predict(X_test)
Y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(Y_test,Y_pred)
```

```python
new_data = np.array([[5,0,14,0,2,3]])
predicted_price = model.predict(new_data)

print(f"Predicted Insurance Price: {predicted_price[0]}")
```
